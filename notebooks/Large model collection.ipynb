```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import numpy as np
```

```python
sns.set_style('white')
sns.set_palette('deep')
```

### Load model summary (pre-computed)

```python
with open('../models/embl_gems/summary.json', 'r') as f:
    data = json.load(f)

genes, reactions, metabolites = data['genes'], data['reactions'], data['metabolites']
```

```python
stats1 = {}
for species in reactions.keys():
    total_rxns = len({r_id for r_id in reactions[species] if not r_id.startswith('R_EX')})
    total_mets = len({m_id[:-2] for m_id in metabolites[species]})
    stats1[species] = (len(genes[species]), total_rxns, total_mets)

df = pd.DataFrame(stats1, index=['genes', 'reactions', 'metabolites']).T
```

### Compare genes vs reactions vs metabolites

```python
g = sns.JointGrid(x='genes', y='reactions', data=df, ratio=4, size=4, xlim=(0,1600), ylim=(0,2500))
g = g.plot_joint(plt.scatter, color=sns.color_palette()[0], alpha=0.1, s=25, edgecolor="white", linewidth=0.5)
g = g.plot_marginals(sns.kdeplot, shade=True, bw=100)
g.ax_marg_x.axes.grid(False)
g.ax_marg_y.axes.grid(False)
g.ax_joint.set_xlabel('number of metabolic genes')
g.ax_joint.set_ylabel('number of reactions')
g.ax_joint.set_xticks([0, 400, 800, 1200, 1600])
plt.savefig('../results/other/embl_gems_1.png', dpi=300)
```

```python
g = sns.JointGrid(x='reactions', y='metabolites', data=df, ratio=4, size=4, xlim=(0,2500), ylim=(0,1200))
g = g.plot_joint(plt.scatter, color=sns.color_palette()[0],
                 alpha=0.2, s=25, edgecolor="white", linewidth=0.5)
g = g.plot_marginals(sns.kdeplot, shade=True, bw=100, color=sns.color_palette('deep')[0])
g.ax_marg_x.axes.grid(False)
g.ax_marg_y.axes.grid(False)
g.ax_joint.set_xlabel('number of reactions')
g.ax_joint.set_ylabel('number of metabolites')
plt.savefig('../results/other/embl_gems_2.png', dpi=300)
```

```python
g = sns.JointGrid(x='genes', y='metabolites', data=df, ratio=4, size=4, xlim=(0,1600), ylim=(0,1200))
g = g.plot_joint(plt.scatter, color=sns.color_palette()[0], alpha=0.1, s=25, edgecolor="white", linewidth=0.5)
g = g.plot_marginals(sns.kdeplot, shade=True, bw=100)
g.ax_marg_x.axes.grid(False)
g.ax_marg_y.axes.grid(False)
g.ax_joint.set_xlabel('number of metabolic genes')
g.ax_joint.set_ylabel('number of metabolites')
g.ax_joint.set_xticks([0, 400, 800, 1200, 1600])
plt.savefig('../results/other/embl_gems_3.png', dpi=300)
```

### Calculate metabolite and reaction frequency 

```python
rxn_cnt = Counter()
for rxns in reactions.values():
    true_rxns = [r_id for r_id in rxns if not r_id.startswith('R_EX')]
    rxn_cnt.update(true_rxns)

df2 = pd.DataFrame(rxn_cnt.items(), columns=['reaction', 'count'])
df2['freq'] = df2['count'] / float(len(reactions))
```

```python
met_cnt = Counter()
for mets in metabolites.values():
    true_mets = list({m_id[:-2] for m_id in mets})
    met_cnt.update(true_mets)

df3 = pd.DataFrame(met_cnt.items(), columns=['metabolite', 'count'])
df3['freq'] = df3['count'] / float(len(metabolites))
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax2 = ax.twinx()

sns.distplot(df2['freq'], hist=True, bins=10, kde=False, norm_hist=False,
             hist_kws={'alpha':0.5, 'color': sns.color_palette()[0]}, ax=ax)

sns.distplot(df3['freq'], hist=True, bins=10, kde=False, norm_hist=False,
             hist_kws={'alpha':0.5, 'color': sns.color_palette()[2]}, ax=ax2)

ax.set_xlim(0,1)
ax.set_ylim(0,1000)
ax.set_xlabel('Frequency')
ax.set_ylabel('Number of reactions')

ax2.set_ylim(0,400)
ax2.set_yticks([0, 100, 200, 300, 400])
ax2.set_ylabel('Number of metabolites')

plt.tight_layout()
plt.savefig('../results/other/embl_gems_4.png', dpi=300)
```

### Plot Metabolic Interaction Potential

```python
df4 = pd.read_csv('../models/embl_gems/community_mip.csv')
df4['normalized'] = df4['mip'] / df4['size']
```

```python
plt.figure(figsize=(7, 3))
ax = sns.boxplot(data=df4, x='size', y='mip', width=0.8, showfliers=False, 
                 color=sns.color_palette()[0], boxprops={'alpha': 0.7} )
ax.set_ylim(0,40)
ax.set_xlabel('Community size')
ax.set_ylabel('Metabolic Interaction Potential')
ax.set_yticks([0, 10, 20, 30, 40])
ax2 = ax.twinx()

sns.pointplot(data=df4, x='size', y='normalized', scale=1, errwidth=2, ax=ax2,
              color=sns.color_palette()[2], join=False)
ax2.set_ylabel('MIP / size')
ax2.set_ylim(1,1.6)
ax2.set_yticks([1, 1.2, 1.4, 1.6])
ax2.grid(False)

plt.tight_layout()
plt.savefig('../results/other/embl_gems_5.png', dpi=300)
```
